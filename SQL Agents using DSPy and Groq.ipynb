````python
import pandas as pd
import random

def generate_random_name():
    # Sample names with more diversity and capitalization
    first_names = [
        'John', 'Jane', 'Mike', 'Emily', 'David', 'Emma', 'Chris', 'Laura', 
        'Daniel', 'Sophia', 'Rob', 'Liz', 'Alex', 'Sam', 'Kate', 'Will', 
        'Ben', 'Mia', 'Olivia', 'Luke', 'Max', 'Ella', 'Leo', 'Chloe'
    ]
    last_names = [
        'Doe', 'Smith', 'Johnson', 'Davis', 'Brown', 'Wilson', 'Miller', 'Moore', 
        'Taylor', 'Anderson', 'Williams', 'Thomas', 'Jackson', 'White', 
        'Harris', 'Martin', 'Garcia', 'Clark', 'Lewis', 'Lee'
    ]
    
    # Randomly select a first and last name
    first_name = random.choice(first_names)
    last_name = random.choice(last_names)
    
    # Randomly apply short forms, capitalization, or initials
    if random.random() > 0.7:  # 30% chance to use a short form or initials
        first_name = first_name[:random.randint(1, 3)].capitalize() + '.'  # Short form
    if random.random() > 0.7:  # 30% chance to use last name initial
        last_name = last_name[0].upper() + '.'
    elif random.random() > 0.8:  # 20% chance to make last name lowercase
        last_name = last_name.lower()
    
    return f"{first_name} {last_name}"

def generate_people_details(employee_ids):
    companies = ['TechCorp', 'InnoSoft', 'DataWorks', 'NetSolutions', 'BizWare']
    
    # Generate people_details data
    people_details_data = {
        'EmployeeID': employee_ids,
        'EmployeeName': [generate_random_name() for _ in employee_ids],
        'CompanyName': [random.choice(companies) for _ in employee_ids],
        'SalaryInLacs': [random.randint(5, 40) for _ in employee_ids]  # Random salary between 5 and 20 lakhs
    }

    return pd.DataFrame(people_details_data)

# Assuming the employee IDs in the employees table are 1 through n
employee_ids = list(range(1, len(df)+1))  # Example: 100 records

# Generate the people_details DataFrame
people_details_df = generate_people_details(employee_ids)

people_details_df[:10]
````

````python
import dspy

class InputSignatureAgent1(dspy.Signature):
    """You are a Senior SQL Programmer who can understand the complex table structures and form SQL queries as per the user's question.
    IMPORTANT instructions:
    1) Dont use JOIN if it is not necessary, check the table and its usable columns, if using JOIN do it with extra CAUTION
    2) Dont forget to apply inverted commas when required.
    3) Give no explanation and directly provide answers.
    4) Take care when joining tables, use right set of table-column configuration"""

    context = dspy.InputField(desc="Will contain the table names and its columns names, data types, can be multiple tables")
    question = dspy.InputField()
    answer = dspy.OutputField(desc="SQL query that will work always, without syntax errors")
    
class ChainOfThoughtSqlProgrammerAgent(dspy.Module):
    def __init__(self):
        super().__init__()
        self.generate_answer = dspy.ChainOfThought(InputSignatureAgent1)

    def forward(self, question):
        context = column_context_employees + '\n'+ column_context_employee_details
        # print(f"Context fetched: {context}")
        prediction = self.generate_answer(context=context, question=question)
        return dspy.Prediction(context=context, answer=prediction.answer)
    
class InputSignatureAgent2(dspy.Signature):
    """PROVIDE NO EXPLANATIONs, You are a SQL Developer lead who can understand the complex table structures and correct SQL queries as per the table structure.
    IMPORTANT instructions:
    1) Remove markdown like ```sql  ```
    2) Check syntax issues in SQL query like removing unwanted characters especially in SELECT COUNT function. 
    3) Dont forget to apply inverted commas when required.
    4) Take care when joining tables, use right set of table-column configuration."""

    context = dspy.InputField(desc="Will contain the table name and its columns names, unique values, data types, can be multiple tables")
    question = dspy.InputField()
    answer = dspy.OutputField(desc="NO EXPLANATION ONLY SQL query that will go into a SQL Compiler so no explanation or other nonsense")
    
class ChainOfThoughtSqlLeadAgent(dspy.Module):
    def __init__(self):
        super().__init__()
        self.generate_answer = dspy.ChainOfThought(InputSignatureAgent2)

    def forward(self, question):
        context = column_context_employees + '\n'+ column_context_employee_details 
        prediction = self.generate_answer(context=context, question=question)
        return dspy.Prediction(context=context, answer=prediction.answer)
````

````python
dev_agent = ChainOfThoughtSqlProgrammerAgent()
sql_correct_agent = ChainOfThoughtSqlLeadAgent()
````

````python
from sqlalchemy import text

def sql_results(query):
    with engine.connect() as conn:
        ans=conn.execute(text(query)).fetchall()       
    return ans
````
